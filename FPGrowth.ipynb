```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
my_data=pd.read_csv("D:\examples\pes2021-all-players_modified_AR.csv")
```

```python
my_data
```

```python
my_data=my_data.drop('Unnamed: 0', axis=1)
```

```python
my_data.head()
```

```python
from mlxtend.frequent_patterns import fpgrowth

frequent_skillsets=fpgrowth(my_data, min_support=0.02,use_colnames=True)
frequent_skillsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
res=association_rules(frequent_skillsets, metric="confidence", min_threshold=0.4)
res=res[['antecedents','consequents','confidence']]
res = res.sort_values(by="confidence", ascending=False)
res
```

```python
import re

for index, row in res.iterrows():
    antecedents = ', '.join([re.sub(r'skill_', '', item).replace('_', ' ').title() for item in row['antecedents']])
    consequents = ', '.join([re.sub(r'skill_', '', item).replace('_', ' ').title() for item in row['consequents']])
    confidence = round(row['confidence'], 3)
    print(f"{antecedents} -> {consequents}. Confidence = {confidence}")
```
